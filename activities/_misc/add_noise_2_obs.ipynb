```python
import os
import numpy as np
import pandas as pd
import pyemu
```

```python
noise_variance = 0.1 #length units
```

```python
true_vals = pd.read_csv(os.path.join("..","FreybergModelForClass","Freyberg_Truth","freyberg.heads"),delim_whitespace=True)
```

```python
true_vals.loc[:,"obsname"] = true_vals.obsname.apply(lambda x: x[6:])
true_vals.index = true_vals.obsname
fore_idx = true_vals.groupby(lambda x:'f' in x).groups[True]
true_vals.loc[fore_idx,"value"] = -999999
true_vals
```

```python
pst = pyemu.Pst(os.path.join("..","FreybergModelForClass","Freyberg_K_only","freyberg.pst"))
pst.observation_data
```

```python
pst.observation_data.loc[true_vals.obsname,"obsval"] = true_vals.value
pst.observation_data
```

```python
pst.write(os.path.join("..","FreybergModelForClass","Freyberg_K_only","freyberg_true_obs.pst"))
```

```python
noise = np.random.standard_normal(true_vals.shape[0])
true_vals.loc[:,"value"] += noise
true_vals.index = true_vals.obsname
fore_idx = true_vals.groupby(lambda x:'f' in x).groups[True]
true_vals.loc[fore_idx,"value"] = -999999
```
