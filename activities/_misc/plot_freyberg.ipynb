```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
TRUTH_HK_PATH = os.path.join("..","FreybergModelForClass","Freyberg_Truth","hk.truth.ref")
def plot_results(model_dir):
    ibnd = np.loadtxt(os.path.join(model_dir,"ibound.ref"))
    if os.path.exists(os.path.join(model_dir,"hk.ref")):
        hk = np.loadtxt(os.path.join(model_dir,"hk.ref"))
    else:
        hk = np.zeros_like(ibnd)
        
    
    hk_t = np.loadtxt(TRUTH_HK_PATH)
    vmin,vmax = min(hk.min(),hk_t.min()),max(hk.max(),hk_t.max())
    #vmin,vmax = None,None
    
    hk[ibnd==0] = np.NaN
    hk_t[ibnd==0] = np.NaN 
    
    fig = plt.figure(figsize=(10,10))
    ax1 = plt.axes((0.05,0.05,0.4,0.8))
    ax1.set_title("best fit")
    ax2 = plt.axes((0.48,0.05,0.4,0.8))
    ax2.set_title("truth")
    axc = plt.axes((0.9,0.05,0.01,0.8))
    ax1.imshow(hk,interpolation="none",vmin=vmin,vmax=vmax)
    a = ax2.imshow(hk_t,interpolation="none",vmin=vmin,vmax=vmax)
    fig.colorbar(a,cax=axc,label="hk")
    
    return fig,ax1,ax2
```

```python
model_ws = os.path.join("..","FreybergModelForClass","Freyberg_k_only")
plot_results(model_ws)
plt.show()
```

```python
model_ws = os.path.join("..","FreybergModelForClass","Freyberg_pilotpoints_master")
plot_results(model_ws)
plt.show()
```
